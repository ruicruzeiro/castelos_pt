This notebook gets the coordinates of every Portuguese castle on the castle dictionary created in notebook 01. It was created as a part of my personal portfolio.

```python
import re
import requests
import time                   # To sleep between each API request
from api_key import api_key   # I used my private Google API key
from castelos_dict import castelos_info
```

```python
count_all = 0
count_missing = 0

for castelo, info in castelos_info.items():
    
    count_all += 1
    
    # GET request
    castelo_url = 'castelo+' + '+'.join(list(castelo.split())) + '+portugal'
    url = 'https://maps.googleapis.com/maps/api/geocode/json?address=' + castelo_url + ',+CA&key=' + api_key
    response = requests.get(url).json()
    
    # Store latitude and longitude from response json into island dictionary    
    if response['results'] != []:
        lat = response['results'][0]['geometry']['location']['lat']
        lon = response['results'][0]['geometry']['location']['lng']
        info['coordinates'] = [lat, lon]
        print(str(count_all) + ' - coordinates added: ' + castelo)
    else:
        print(str(count_all) + ' - coordinates missing: ' + castelo)
        count_missing += 1
    
    # Keeping the API happy
    time.sleep(1)

print('')
print('***DONE***')
print('Coordinates added:', count_all - count_missing)
print('Coordinates missing:', count_missing)
print('Total castles: ', count_all)
```
