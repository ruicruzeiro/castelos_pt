This notebook generates a map with every Portuguese castle on the castle dictionary created in notebooks 01 and 02. It was created as a part of my personal portfolio.

```python
import folium as fl
from castelos_dict import castelos_info
```

```python
name_list = []
lat_list = []
lon_list = []

for castelo, info in castelos_info.items():
    if type(info['coordinates']) == list:
        name_list.append(castelo)
        lat_list.append(info['coordinates'][0])
        lon_list.append(info['coordinates'][1])
```

```python
icon_castelo = fl.features.CustomIcon('icon_castelo.jpeg', icon_size=(50,50))
```

```python
map = fl.Map(location=[38.283496, -17.667667], tiles='cartodbpositron', zoom_start=5)
feature_group = fl.FeatureGroup("Castelos")

for castle_num in range(len(name_list)):
    iframe = fl.IFrame('<strong>' + name_list[castle_num] + '</strong>')
    popup = fl.Popup(iframe, min_width=300, max_width=300, max_height=100)
    fl.Marker(location=[lat_list[castle_num], lon_list[castle_num]], popup=name_list[castle_num], icon=icon_castelo).add_to(feature_group)

feature_group.add_to(map)
fl.LayerControl().add_to(map)

map
```

```python
map = fl.Map(location=[39.9185272, -7.9137986], tiles='cartodbpositron', zoom_start=7)
feature_group = fl.FeatureGroup("Castelos")

for lat, lon, name in zip(lat_list, lon_list, name_list):
    iframe = fl.IFrame('<strong>' + name + '</strong>')
    popup = fl.Popup(iframe, min_width=300, max_width=300, max_height=100)
    feature_group.add_child(fl.Marker(location=[lat,lon], popup=popup, icon=icon_castelo))
    
map.add_child(feature_group)

map
```
